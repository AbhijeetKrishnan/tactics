```python
import matplotlib.pyplot as plt
```

```python
tactics = [
    'can_threat',
    'can_check',
    'can_fork',
    'discovered_check',
    'discovered_threat',
    'skewer',
    'pin',
    'random'
]

coverage = [0.96, 0.45, 0.32, 0, 0.96, 0.22, 0.79, 1.00]
divergence_sf14 = [378.94, 549.19, 676.45, 338.55, 375.97, 748.4, 526.45, 328.09]
divergence_m1100 = [9.22, 4.02, 4.67, 18.64, 1.19, 5.41, 4.9, 8.28]
```

```python
sorted_coverage = sorted(zip(coverage, tactics))
coverage, tactics = list(zip(*sorted_coverage))
plt.bar(range(len(tactics)), coverage, tick_label=tactics, color=['tab:blue', 'tab:blue', 'tab:blue', 'tab:blue', 'tab:blue', 'tab:blue', 'tab:blue', 'tab:red'])
plt.xticks(range(len(tactics)), tactics, rotation='vertical')
plt.show()
```

```python
sorted_divergence = sorted(zip(divergence_sf14, tactics))
divergence, tactics = list(zip(*sorted_divergence))
plt.bar(range(len(tactics)), divergence, tick_label=tactics, color=['tab:red', 'tab:blue', 'tab:blue', 'tab:blue', 'tab:blue', 'tab:blue', 'tab:blue', 'tab:blue'])
plt.xticks(range(len(tactics)), tactics, rotation='vertical')
plt.show()
```

```python
sorted_divergence = sorted(zip(divergence_m1100, tactics))
divergence, tactics = list(zip(*sorted_divergence))
plt.bar(range(len(tactics)), divergence, tick_label=tactics, color=['tab:blue', 'tab:blue', 'tab:blue', 'tab:blue', 'tab:blue', 'tab:red', 'tab:blue', 'tab:blue'])
plt.xticks(range(len(tactics)), tactics, rotation='vertical')
plt.show()
```
